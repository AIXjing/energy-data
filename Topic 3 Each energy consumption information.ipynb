```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('energy_cons.csv')
```

```python
df.head(10)
```

```python
# Select data from 2019 (the latest year) and dropna(), which means to take away all the rows that contain NA value.
df_2019 = df.loc[df['year'] == 2019].dropna()

df_2019.head(10)
```

```python
df_2019['nonrenewable_energy'] = df_2019.iloc[:,3:6].sum(axis=1)

df_2019['total']=df_2019.iloc[:,13] + df_2019.iloc[:,16]

df_2019.head(10)
```

```python
import sqlite3
con_cons_2019 = sqlite3.connect('cons_2019.db')
df_2019.to_sql('cons_2019', con_cons_2019)
```

```python
%%sql
SELECT *
FROM 'cons_2019'
LIMIT 10
```

```python
%%sql
SELECT COUNT(DISTINCT(iso_code)), COUNT(*)
FROM 'cons_2019'
```

```python
%%sql

SELECT country, fossil_fuel_consumption
FROM 'cons_2019'
WHERE country != 'World'
ORDER BY fossil_fuel_consumption DESC
LIMIT 10
```

```python
df_2019_fuel_top10 = df_2019.sort_values(by=['fossil_fuel_consumption'], ascending=False).iloc[1:10,]
```

```python
df_2019_fuel_top10.plot.bar(
    'country',
    ['fossil_fuel_consumption','low_carbon_consumption'],
    rot=30,
    figsize=(9,6),
    width=0.75
)

plt.title('Fossil fuel consumption TOP 10 countries', fontsize=18)
plt.legend(['Fossil Fule', 'Renewables'], fontsize=12)
plt.xlabel('')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
